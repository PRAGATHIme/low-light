```python
import os, random, shutil
from collections import defaultdict

# ✅ Correct double-nested path
ANNOT_BASE   = '/content/ExDark_Annno/ExDark_Annno'
IMAGES_BASE  = '/content/drive/MyDrive/archive'   # adjust to your images root
OUTPUT_DIR   = '/content/balanced_dataset'
TARGET_TOTAL = 1500
TRAIN_RATIO  = 0.8
RANDOM_SEED  = 42

# ✅ Class name → YOLO integer id (sorted alphabetically to match folder names)
CLASS_NAMES = ['Bicycle','Boat','Bottle','Bus','Car',
               'Cat','Chair','Cup','Dog','Motorbike','People','Table']
CLASS_TO_ID = {c: i for i, c in enumerate(CLASS_NAMES)}

print("✅ Class map:")
for name, idx in CLASS_TO_ID.items():
    print(f"   {idx}: {name}")
```

```python
def parse_bbgt_to_yolo(txt_path, img_w, img_h):
    """
    Reads ExDark bbGt annotation and returns YOLO format lines.
    bbGt format:  ClassName x y w h 0 0 0 0 0 0 0
    YOLO format:  class_id cx cy w h   (all normalized 0-1)
    """
    yolo_lines = []
    with open(txt_path, 'r', encoding='latin-1', errors='ignore') as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith('%'):   # skip header
                continue
            parts = line.split()
            if len(parts) < 5:
                continue
            cls_name = parts[0]
            if cls_name not in CLASS_TO_ID:
                continue
            try:
                x  = float(parts[1])   # top-left x
                y  = float(parts[2])   # top-left y
                w  = float(parts[3])   # width
                h  = float(parts[4])   # height
            except ValueError:
                continue

            # Convert to YOLO normalized center format
            cx = (x + w / 2) / img_w
            cy = (y + h / 2) / img_h
            nw = w / img_w
            nh = h / img_h

            # Clamp to [0,1]
            cx = max(0, min(1, cx))
            cy = max(0, min(1, cy))
            nw = max(0, min(1, nw))
            nh = max(0, min(1, nh))

            cls_id = CLASS_TO_ID[cls_name]
            yolo_lines.append(f"{cls_id} {cx:.6f} {cy:.6f} {nw:.6f} {nh:.6f}")
    return yolo_lines

# ✅ Test on the sample file we saw
import cv2
sample_txt = '/content/ExDark_Annno/ExDark_Annno/Chair/2015_03777.png.txt'
print("Raw content:")
with open(sample_txt, encoding='latin-1') as f:
    print(f.read())

# We need image size to normalize — use a dummy 640x480 just to test parser
lines = parse_bbgt_to_yolo(sample_txt, img_w=640, img_h=480)
print("\nConverted YOLO lines (with dummy img size):")
for l in lines:
    print(" ", l)
```

```python
class_to_images = defaultdict(list)

for cls in CLASS_NAMES:
    cls_dir = os.path.join(ANNOT_BASE, cls)
    if not os.path.exists(cls_dir):
        print(f"⚠️ Missing folder: {cls_dir}")
        continue
    for txt_file in os.listdir(cls_dir):
        if txt_file.startswith('._') or txt_file.startswith('.'):
            continue
        if not txt_file.endswith('.txt'):
            continue
        class_to_images[cls].append(txt_file)  # e.g. "2015_03777.png.txt"

print("📊 Images per class:")
for cls in CLASS_NAMES:
    print(f"   {cls:12s}: {len(class_to_images[cls])}")

per_class = TARGET_TOTAL // len(CLASS_NAMES)
print(f"\n⚖️  {per_class} per class × {len(CLASS_NAMES)} = {per_class*len(CLASS_NAMES)} total")

random.seed(RANDOM_SEED)
selected_by_class = {}
all_selected = []  # list of (cls, txt_filename)

for cls in CLASS_NAMES:
    imgs = class_to_images[cls]
    n = min(per_class, len(imgs))
    sampled = random.sample(imgs, n)
    selected_by_class[cls] = sampled
    all_selected.extend([(cls, f) for f in sampled])
    print(f"   {cls:12s}: {n} selected")

random.shuffle(all_selected)
print(f"\n✅ Total selected: {len(all_selected)}")
```

```python
# ✅ SET THIS after Cell 4 tells you where images are
# e.g. IMAGES_BASE = '/content/drive/MyDrive/archive/ExDark'
IMAGES_BASE = '/content/drive/MyDrive/archive'   # ← update after Cell 4

split_idx = int(len(all_selected) * TRAIN_RATIO)
train_set = all_selected[:split_idx]
val_set   = all_selected[split_idx:]

def process_split(dataset, split_name):
    img_out = os.path.join(OUTPUT_DIR, split_name, 'images')
    lbl_out = os.path.join(OUTPUT_DIR, split_name, 'labels')
    os.makedirs(img_out, exist_ok=True)
    os.makedirs(lbl_out, exist_ok=True)

    ok, miss_img, miss_lbl, empty_lbl = 0, 0, 0, 0

    for cls, txt_file in dataset:
        # txt_file = "2015_03777.png.txt"
        img_name = txt_file[:-4]          # "2015_03777.png"
        base     = img_name.rsplit('.', 1)[0]  # "2015_03777"

        # --- Find image ---
        img_src = None
        for folder in [cls, '']:
            for ext in [img_name] + [f"{base}.{e}" for e in ['jpg','jpeg','png','JPG','JPEG','PNG']]:
                candidate = os.path.join(IMAGES_BASE, folder, ext)
                if os.path.exists(candidate):
                    img_src = candidate
                    break
            if img_src: break

        if img_src:
            dst_img = os.path.join(img_out, os.path.basename(img_src))
            shutil.copy(img_src, dst_img)
            # Get real image dimensions for accurate YOLO conversion
            import cv2
            img = cv2.imread(img_src)
            if img is not None:
                img_h, img_w = img.shape[:2]
            else:
                img_h, img_w = 640, 640   # fallback
        else:
            miss_img += 1
            img_h, img_w = 640, 640       # fallback size

        # --- Convert & save annotation ---
        ann_src = os.path.join(ANNOT_BASE, cls, txt_file)
        if os.path.exists(ann_src):
            yolo_lines = parse_bbgt_to_yolo(ann_src, img_w, img_h)
            if yolo_lines:
                with open(os.path.join(lbl_out, base + '.txt'), 'w') as f:
                    f.write('\n'.join(yolo_lines))
                ok += 1
            else:
                empty_lbl += 1
        else:
            miss_lbl += 1

    print(f"  {split_name:6s} → ✅ converted: {ok} | "
          f"⚠️ miss img: {miss_img} | "
          f"⚠️ miss lbl: {miss_lbl} | "
          f"⚠️ empty lbl: {empty_lbl}")

print("📁 Processing splits...")
process_split(train_set, 'train')
process_split(val_set,   'val')
print("✅ Done!")
```

```python
def enhance_gamma_clahe(img_path, out_path, gamma=1.8, clip=2.0):
    img = cv2.imread(str(img_path))
    if img is None: return False

    # Step 1: gamma lift — brightens globally without crushing blacks
    inv = 1.0 / gamma
    table = (np.arange(256) / 255.0) ** inv * 255
    img = cv2.LUT(img, table.astype(np.uint8))

    # Step 2: gentle CLAHE for local contrast (clip=2.0, not 3.0)
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=clip, tileGridSize=(8, 8))
    l = clahe.apply(l)
    img = cv2.cvtColor(cv2.merge((l, a, b)), cv2.COLOR_LAB2BGR)

    cv2.imwrite(str(out_path), img)
    return True
```

```python
import cv2, numpy as np, os, shutil, yaml
from pathlib import Path

# 1. Point to the actual folder we built
DARK_DIR     = '/content/yolo_dataset'
ENHANCED_DIR = '/content/enhanced_gamma_clahe'

def enhance_gamma_clahe(img_path, out_path, gamma=1.8, clip=2.0):
    img = cv2.imread(str(img_path))
    if img is None:
        return False

    # Step 1: Gamma lift — raises overall brightness softly
    inv_gamma = 1.0 / gamma
    table     = (np.arange(256) / 255.0) ** inv_gamma * 255
    img       = cv2.LUT(img, table.astype(np.uint8))

    # Step 2: Gentle CLAHE for local contrast (lower clip than before)
    lab      = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b  = cv2.split(lab)
    clahe    = cv2.createCLAHE(clipLimit=clip, tileGridSize=(8, 8))
    l        = clahe.apply(l)
    enhanced = cv2.cvtColor(cv2.merge((l, a, b)), cv2.COLOR_LAB2BGR)

    cv2.imwrite(str(out_path), enhanced)
    return True

# ---- Process all splits ----
total, failed = 0, 0

for split in ['train', 'val']:
    # 2. Corrected YOLO structure: images/split
    in_img  = Path(f'{DARK_DIR}/images/{split}')

    # Safety check: skip the split if the folder doesn't exist
    if not in_img.exists():
        print(f"⏭️ Skipping '{split}' split because {in_img} does not exist.")
        continue

    out_img = Path(f'{ENHANCED_DIR}/images/{split}')
    out_img.mkdir(parents=True, exist_ok=True)

    for img_file in in_img.iterdir():
        if img_file.suffix.lower() not in ['.jpg','.jpeg','.png','.bmp']:
            continue
        if enhance_gamma_clahe(img_file, out_img / img_file.name):
            total += 1
        else:
            failed += 1

    # 3. Corrected label path structure: labels/split
    in_labels = Path(f'{DARK_DIR}/labels/{split}')
    if in_labels.exists():
        shutil.copytree(
            str(in_labels),
            f'{ENHANCED_DIR}/labels/{split}',
            dirs_exist_ok=True
        )
    print(f'  ✅ {split}: done')

print(f'\n✅ Enhanced: {total} images | ❌ Failed: {failed}')

# 4. Generate the YAML file so it's ready for training
cfg = {
    'path'  : ENHANCED_DIR,
    'train' : 'images/train',
    'val'   : 'images/val' if Path(f'{ENHANCED_DIR}/images/val').exists() else 'images/train',
    'nc'    : 12,
    'names' : ['Bicycle','Boat','Bottle','Bus','Car',
               'Cat','Chair','Cup','Dog','Motorbike','People','Table']
}

yaml_path = Path(f'{ENHANCED_DIR}/dataset.yaml')
with open(yaml_path, 'w') as f:
    yaml.dump(cfg, f, default_flow_style=False)
print("✅ dataset.yaml created for Gamma+CLAHE dataset")
```

```python
import os, shutil, random, yaml, cv2
from collections import defaultdict

# ---- Paths ----
ANNOT_BASE   = '/content/ExDark_Annno/ExDark_Annno'
IMAGES_BASE  = '/content/drive/MyDrive/archive'
OUTPUT_DIR   = '/content/balanced_dataset'
TARGET_TOTAL = 1500
TRAIN_RATIO  = 0.8
RANDOM_SEED  = 42
CLASS_NAMES  = ['Bicycle','Boat','Bottle','Bus','Car',
                'Cat','Chair','Cup','Dog','Motorbike','People','Table']
CLASS_TO_ID  = {c:i for i,c in enumerate(CLASS_NAMES)}

def parse_bbgt_to_yolo(txt_path, img_w, img_h):
    yolo_lines = []
    with open(txt_path, 'r', encoding='latin-1', errors='ignore') as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith('%'):
                continue
            parts = line.split()
            if len(parts) < 5:
                continue
            cls_name = parts[0]
            if cls_name not in CLASS_TO_ID:
                continue
            try:
                x, y, w, h = float(parts[1]), float(parts[2]), float(parts[3]), float(parts[4])
            except ValueError:
                continue

            # YOLO Normalization Math
            cx = max(0, min(1, (x + w/2) / img_w))
            cy = max(0, min(1, (y + h/2) / img_h))
            nw = max(0, min(1, w / img_w))
            nh = max(0, min(1, h / img_h))
            yolo_lines.append(f"{CLASS_TO_ID[cls_name]} {cx:.6f} {cy:.6f} {nw:.6f} {nh:.6f}")
    return yolo_lines

# --- 1. Gather & Sample Images ---
class_to_images = defaultdict(list)
for cls in CLASS_NAMES:
    cls_dir = os.path.join(ANNOT_BASE, cls)
    if not os.path.exists(cls_dir):
        print(f"⚠️ Missing: {cls_dir}")
        continue
    for f in os.listdir(cls_dir):
        if not f.startswith('._') and not f.startswith('.') and f.endswith('.txt'):
            class_to_images[cls].append(f)

per_class = TARGET_TOTAL // len(CLASS_NAMES)
random.seed(RANDOM_SEED)
all_selected = []
for cls in CLASS_NAMES:
    imgs = class_to_images[cls]
    sampled = random.sample(imgs, min(per_class, len(imgs)))
    all_selected.extend([(cls, f) for f in sampled])
    print(f"  {cls:12s}: {len(sampled)}")

random.shuffle(all_selected)
split_idx  = int(len(all_selected) * TRAIN_RATIO)
train_set  = all_selected[:split_idx]
val_set    = all_selected[split_idx:]
print(f"\n✅ Total: {len(all_selected)} | Train: {len(train_set)} | Val: {len(val_set)}")

# --- 2. Copy Images & Convert Labels ---
def process_split(dataset, split_name):
    # Standardized YOLO Pathing
    img_out = os.path.join(OUTPUT_DIR, 'images', split_name)
    lbl_out = os.path.join(OUTPUT_DIR, 'labels', split_name)
    os.makedirs(img_out, exist_ok=True)
    os.makedirs(lbl_out, exist_ok=True)

    ok, miss_img = 0, 0
    for cls, txt_file in dataset:
        img_name = txt_file[:-4]
        base     = img_name.rsplit('.', 1)[0]
        img_src  = None

        # Hunt for the image across possible extensions
        for folder in [cls, '']:
            for ext in [img_name] + [f"{base}.{e}" for e in ['jpg','jpeg','png','JPG','JPEG','PNG']]:
                c = os.path.join(IMAGES_BASE, folder, ext)
                if os.path.exists(c):
                    img_src = c; break
            if img_src: break

        if img_src:
            img = cv2.imread(img_src)
            # Safe Check: Skip if image is corrupt to avoid breaking YOLO math
            if img is None:
                miss_img += 1
                continue

            ih, iw = img.shape[:2]
            ann_src = os.path.join(ANNOT_BASE, cls, txt_file)

            if os.path.exists(ann_src):
                lines = parse_bbgt_to_yolo(ann_src, iw, ih)
                if lines:
                    # Write label file
                    with open(os.path.join(lbl_out, base+'.txt'), 'w') as f:
                        f.write('\n'.join(lines))
                    # Copy image only if label was successfully created
                    shutil.copy(img_src, os.path.join(img_out, os.path.basename(img_src)))
                    ok += 1
        else:
            miss_img += 1

    print(f"  {split_name}: ✅ {ok} converted | ⚠️ {miss_img} missing or corrupt imgs")

# Clean up any ghost folders from a previous run
if os.path.exists(OUTPUT_DIR):
    shutil.rmtree(OUTPUT_DIR)

print("\n📁 Building balanced dataset...")
process_split(train_set, 'train')
process_split(val_set,   'val')

# --- 3. Write Config YAML ---
cfg = {
    'path': OUTPUT_DIR,
    'train': 'images/train',
    'val': 'images/val',
    'nc': 12,
    'names': CLASS_NAMES
}

with open(f'{OUTPUT_DIR}/dataset.yaml','w') as f:
    yaml.dump(cfg, f, default_flow_style=False)
print(f"✅ dataset.yaml written safely to {OUTPUT_DIR}/dataset.yaml")
```

```python
import torch
import torch.nn as nn

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"✅ PyTorch {torch.__version__} | Device: {DEVICE}")

class ZeroDCE(nn.Module):
    def __init__(self):
        super().__init__()
        self.relu    = nn.ReLU(inplace=True)
        self.e_conv1 = nn.Conv2d(3,  32, 3, 1, 1, bias=True)
        self.e_conv2 = nn.Conv2d(32, 32, 3, 1, 1, bias=True)
        self.e_conv3 = nn.Conv2d(32, 32, 3, 1, 1, bias=True)
        self.e_conv4 = nn.Conv2d(32, 32, 3, 1, 1, bias=True)
        self.e_conv5 = nn.Conv2d(64, 32, 3, 1, 1, bias=True)
        self.e_conv6 = nn.Conv2d(64, 32, 3, 1, 1, bias=True)
        self.e_conv7 = nn.Conv2d(64, 24, 3, 1, 1, bias=True)

    def forward(self, x):
        x1 = self.relu(self.e_conv1(x))
        x2 = self.relu(self.e_conv2(x1))
        x3 = self.relu(self.e_conv3(x2))
        x4 = self.relu(self.e_conv4(x3))
        x5 = self.relu(self.e_conv5(torch.cat([x3, x4], 1)))
        x6 = self.relu(self.e_conv6(torch.cat([x2, x5], 1)))
        return torch.tanh(self.e_conv7(torch.cat([x1, x6], 1)))

def apply_curves(img, curves):
    r = img
    for curve in torch.split(curves, 3, dim=1):
        r = r + curve * (torch.pow(r, 2) - r)
    return torch.clamp(r, 0, 1)

print("✅ Zero-DCE model defined")
```

```python
import torch, torch.nn as nn, torch.optim as optim
import torchvision.transforms as T
from PIL import Image
from pathlib import Path
import os

# ---- Losses (all fixed) ----
def loss_illumination(enh):
    """Brightness should be well exposed ~0.6"""
    return torch.mean(torch.abs(torch.mean(enh) - 0.6))

def loss_spatial(enh, ori):
    """
    ✅ FIXED: use padding='same' equivalent via padding=1 on a 3×3 kernel
    instead of padding=2 on 4×4 which causes size mismatch
    """
    k  = torch.ones(1, 1, 3, 3).to(DEVICE) / 9.0   # 3×3 averaging kernel
    eg = enh.mean(1, keepdim=True)
    og = ori.mean(1, keepdim=True)
    # padding=1 on 3×3 kernel keeps spatial dims identical ✅
    es  = nn.functional.conv2d(eg, k, padding=1)
    os_ = nn.functional.conv2d(og, k, padding=1)
    return torch.mean(torch.pow((eg - es) - (og - os_), 2))

def loss_color(enh):
    """R,G,B channels should stay balanced"""
    r, g, b = enh[:, 0], enh[:, 1], enh[:, 2]
    return (torch.mean(torch.pow(r - g, 2)) +
            torch.mean(torch.pow(r - b, 2)) +
            torch.mean(torch.pow(g - b, 2)))

def loss_tv(curves):
    """Smooth curve maps — avoid harsh pixel jumps"""
    dy = torch.abs(curves[:, :, 1:, :] - curves[:, :, :-1, :])
    dx = torch.abs(curves[:, :, :, 1:] - curves[:, :, :, :-1])
    return torch.mean(dx) + torch.mean(dy)

# ---- Dataset ----
class DarkDataset(torch.utils.data.Dataset):
    def __init__(self, img_dir):
        self.files = [
            f for f in Path(img_dir).iterdir()
            if f.suffix.lower() in ['.jpg', '.jpeg', '.png']
        ]
        # ✅ Resize to 256×256 (even number — avoids odd-dim conv issues)
        self.tf = T.Compose([
            T.Resize((256, 256)),
            T.ToTensor()
        ])
        print(f"  Found {len(self.files)} images")

    def __len__(self):
        return len(self.files)

    def __getitem__(self, i):
        img = Image.open(self.files[i]).convert('RGB')
        return self.tf(img)

# ✅ Correct path — matches your dataset structure
TRAIN_IMG_DIR = '/content/balanced_dataset/images/train'

dataset    = DarkDataset(TRAIN_IMG_DIR)
dataloader = torch.utils.data.DataLoader(
    dataset,
    batch_size  = 8,
    shuffle     = True,
    num_workers = 2,
    pin_memory  = True
)

# ---- Model + optimizer ----
model_dce = ZeroDCE().to(DEVICE)
optimizer = optim.Adam(model_dce.parameters(), lr=1e-4)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=100)

# ---- Training loop ----
print(f"🚀 Training Zero-DCE for 100 epochs on {DEVICE}...")
print(f"   Batches per epoch: {len(dataloader)}\n")

for epoch in range(100):
    model_dce.train()
    total_loss = 0.0
    l_ill_sum = l_spa_sum = l_col_sum = l_tv_sum = 0.0

    for imgs in dataloader:
        imgs     = imgs.to(DEVICE)
        curves   = model_dce(imgs)
        enhanced = apply_curves(imgs, curves)

        l_ill = loss_illumination(enhanced)  * 1.0
        l_spa = loss_spatial(enhanced, imgs) * 1.0
        l_col = loss_color(enhanced)         * 0.5
        l_tv  = loss_tv(curves)              * 200.0
        loss  = l_ill + l_spa + l_col + l_tv

        optimizer.zero_grad()
        loss.backward()
        # ✅ Gradient clipping — prevents exploding gradients
        torch.nn.utils.clip_grad_norm_(model_dce.parameters(), max_norm=1.0)
        optimizer.step()

        total_loss += loss.item()
        l_ill_sum  += l_ill.item()
        l_spa_sum  += l_spa.item()
        l_col_sum  += l_col.item()
        l_tv_sum   += l_tv.item()

    scheduler.step()

    if (epoch + 1) % 10 == 0:
        n = len(dataloader)
        print(f"  Epoch {epoch+1:3d}/100 | "
              f"loss: {total_loss/n:.4f} | "
              f"ill: {l_ill_sum/n:.3f} | "
              f"spa: {l_spa_sum/n:.3f} | "
              f"col: {l_col_sum/n:.3f} | "
              f"tv: {l_tv_sum/n:.3f} | "
              f"lr: {scheduler.get_last_lr()[0]:.6f}")

# ---- Save weights ----
os.makedirs('/content/zerodce', exist_ok=True)
torch.save(model_dce.state_dict(), '/content/zerodce/weights.pth')
size = os.path.getsize('/content/zerodce/weights.pth')
print(f"\n✅ Zero-DCE trained & saved ({size/1024:.1f} KB)")
print("   → Now run Cell 5 to enhance all images")
```

```python
import torch, torch.nn as nn, torch.optim as optim
import torchvision.transforms as T
from PIL import Image
from pathlib import Path
import os, gc

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'

# ---- Losses ----
def loss_illumination(enh):
    return torch.mean(torch.abs(torch.mean(enh) - 0.6))

def loss_spatial(enh, ori):
    k   = torch.ones(1, 1, 3, 3).to(DEVICE) / 9.0
    eg  = enh.mean(1, keepdim=True)
    og  = ori.mean(1, keepdim=True)
    es  = nn.functional.conv2d(eg, k, padding=1)
    os_ = nn.functional.conv2d(og, k, padding=1)
    return torch.mean(torch.pow((eg - es) - (og - os_), 2))

def loss_color(enh):
    r, g, b = enh[:,0], enh[:,1], enh[:,2]
    return (torch.mean(torch.pow(r-g, 2)) +
            torch.mean(torch.pow(r-b, 2)) +
            torch.mean(torch.pow(g-b, 2)))

def loss_tv(curves):
    dy = torch.abs(curves[:,:,1:,:]  - curves[:,:,:-1,:])
    dx = torch.abs(curves[:,:,:,1:]  - curves[:,:,:,:-1])
    return torch.mean(dx) + torch.mean(dy)

# ---- Dataset ----
class DarkDataset(torch.utils.data.Dataset):
    def __init__(self, img_dir):
        self.files = [
            f for f in Path(img_dir).iterdir()
            if f.suffix.lower() in ['.jpg','.jpeg','.png']
        ]
        # ✅ 128×128 — 4× less memory than 256×256, still enough for Zero-DCE
        self.tf = T.Compose([T.Resize((128, 128)), T.ToTensor()])
        print(f"  Found {len(self.files)} training images")

    def __len__(self): return len(self.files)
    def __getitem__(self, i):
        return self.tf(Image.open(self.files[i]).convert('RGB'))

dataset = DarkDataset('/content/balanced_dataset/images/train')
dataloader = torch.utils.data.DataLoader(
    dataset,
    batch_size  = 4,        # ✅ reduced from 8 → 4
    shuffle     = True,
    num_workers = 1,        # ✅ reduced from 2 → 1
    pin_memory  = True
)

# ---- Model ----
gc.collect()
torch.cuda.empty_cache()

model_dce = ZeroDCE().to(DEVICE)
optimizer = optim.Adam(model_dce.parameters(), lr=1e-4)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=100)

print(f"🚀 Training Zero-DCE — 100 epochs on {DEVICE}")
print(f"   GPU free: {(torch.cuda.get_device_properties(0).total_memory - torch.cuda.memory_allocated())/1024**3:.1f} GB\n")

for epoch in range(100):
    model_dce.train()
    total_loss = 0.0

    for imgs in dataloader:
        imgs     = imgs.to(DEVICE)
        curves   = model_dce(imgs)
        enhanced = apply_curves(imgs, curves)

        loss = (loss_illumination(enhanced) * 1.0  +
                loss_spatial(enhanced, imgs) * 1.0  +
                loss_color(enhanced)         * 0.5  +
                loss_tv(curves)              * 200.0)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_dce.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item()

        # ✅ Free graph after each batch — prevents memory buildup
        del imgs, curves, enhanced, loss
        torch.cuda.empty_cache()

    scheduler.step()

    if (epoch + 1) % 10 == 0:
        n = len(dataloader)
        mem_used = torch.cuda.memory_allocated() / 1024**3
        print(f"  Epoch {epoch+1:3d}/100 | "
              f"loss: {total_loss/n:.4f} | "
              f"lr: {scheduler.get_last_lr()[0]:.6f} | "
              f"GPU: {mem_used:.2f}GB")

os.makedirs('/content/zerodce', exist_ok=True)
torch.save(model_dce.state_dict(), '/content/zerodce/weights.pth')
print(f"\n✅ Zero-DCE saved → /content/zerodce/weights.pth")
```

```python
import os, shutil, cv2, torch, yaml
import numpy as np
from pathlib import Path
import gc # Import garbage collector

# ---- Paths (Ensuring consistency with YOLO structure) ----
OUTPUT_DIR = '/content/balanced_dataset'
ZDCE_DIR   = '/content/enhanced_zerodce'

# 1. Clean up any ghost folders from a previous run
if os.path.exists(ZDCE_DIR):
    shutil.rmtree(ZDCE_DIR)

# Ensure model is in eval mode
try:
    model_dce.eval()
except NameError:
    print("❌ ERROR: 'model_dce' is not defined. Please ensure your Zero-DCE model cell is run first.")



def enhance_zerodce(img_path, out_path):
    img_bgr = cv2.imread(str(img_path))
    if img_bgr is None: return False

    # 1. NEW: Cap image size to prevent OOM explosions
    h, w = img_bgr.shape[:2]
    max_dim = 1280 # Safe limit for 15GB VRAM
    if max(h, w) > max_dim:
        scale = max_dim / max(h, w)
        img_bgr = cv2.resize(img_bgr, (int(w * scale), int(h * scale)))

    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    tensor  = torch.from_numpy(
        img_rgb.astype(np.float32) / 255.0
    ).permute(2, 0, 1).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        curves   = model_dce(tensor)
        enhanced = apply_curves(tensor, curves)

    out_np  = (enhanced.squeeze(0).permute(1, 2, 0).cpu().numpy() * 255).clip(0, 255).astype(np.uint8)

    # 2. NEW: Manually delete variables and flush the GPU cache
    del tensor, curves, enhanced
    torch.cuda.empty_cache()
    gc.collect()

    cv2.imwrite(str(out_path), cv2.cvtColor(out_np, cv2.COLOR_RGB2BGR))
    return True
print("\n✨ Processing images through Zero-DCE Neural Network...")
total, failed, missing_dirs = 0, 0, 0

# 2. Process Train and Val Splits
for split in ['train', 'val']:
    # Standardized YOLO Pathing (images/split)
    in_img  = Path(f'{OUTPUT_DIR}/images/{split}')
    out_img = Path(f'{ZDCE_DIR}/images/{split}')

    if not in_img.exists():
        print(f"⚠️ Warning: Source directory {in_img} not found. Skipping {split}.")
        missing_dirs += 1
        continue

    out_img.mkdir(parents=True, exist_ok=True)
    files = list(in_img.iterdir())

    print(f"\nProcessing {split} split ({len(files)} files)...")
    for i, f in enumerate(files):
        if f.suffix.lower() in ['.jpg', '.jpeg', '.png', '.bmp']:
            if enhance_zerodce(f, out_img / f.name):
                total += 1
            else:
                failed += 1

        # Progress tracker
        if (i + 1) % 200 == 0 or (i + 1) == len(files):
            print(f"  -> {split}: {i + 1}/{len(files)} completed")

    # Standardized YOLO Pathing for labels (labels/split)
    in_lbl  = Path(f'{OUTPUT_DIR}/labels/{split}')
    out_lbl = Path(f'{ZDCE_DIR}/labels/{split}')

    if in_lbl.exists():
        shutil.copytree(str(in_lbl), str(out_lbl), dirs_exist_ok=True)

    print(f"✅ {split.capitalize()} split done!")

if missing_dirs == 2:
    print("\n❌ CRITICAL ERROR: Neither train nor val folders were found in the balanced dataset!")
else:
    print(f"\n✅ Total Enhanced: {total} | ❌ Corrupt/Failed: {failed}")

# 3. Write Config YAML
# Safety check to ensure CLASS_NAMES is in memory
try:
    _ = CLASS_NAMES
except NameError:
    CLASS_NAMES = ['Bicycle','Boat','Bottle','Bus','Car','Cat','Chair','Cup','Dog','Motorbike','People','Table']

cfg = {
    'path': ZDCE_DIR,
    'train': 'images/train', # Fixed YOLO pathing
    'val': 'images/val',     # Fixed YOLO pathing
    'nc': 12,
    'names': CLASS_NAMES
}

yaml_path = f'{ZDCE_DIR}/dataset.yaml'
with open(yaml_path, 'w') as f:
    yaml.dump(cfg, f, default_flow_style=False)
print(f"📄 dataset.yaml written safely to {yaml_path}")
```
